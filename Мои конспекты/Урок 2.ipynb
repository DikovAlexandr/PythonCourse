Задача 2

Напишите программу, создающую два списка, заполненных случайным набором элементов (используются цифры и латинские буквы). Выведите на печать элементы, общие для двух списков. Подсказка: для построения финального списка используйте множества.


```python
def merge(n, m):
  random.seed(version=2)
  s = 'ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz1234567890'
  print(random.sample(s, n))
  lst1 = list(random.sample(s, n))
  print(lst1)
  lst2 = list(random.sample(s, m))
  print(lst2)
  print(set(lst1).intersection(set(lst2)))
```

```python
merge(10, 11)
```

Задача 3

Напишите функцию для рекурсивного вычисления факториалов произвольного натурального числа, n!

```python
def fact(n):
  res = 1
  if n == 0:
    return 1
  res = fact(n-1) * n
  return res
```

```python
def queens(n):
  a = list(range(n))
  up = [True]*(2*n - 1)
  down = [True]*(2*n - 1)


  def sub(i):
    if i == n:
      yield tuple(a)
    else:
      for k in range(i, n):
        j = a[k]
        p = i + j
        q = i - j + n - 1
        if up[p] and down [q]:
          up[p] = down[q] = False 
          a[i], a[k] = a[k], a[i]
          yield from sub(i + 1)
          up[p] = down[q] = True
          a[i], a[k] = a[k], a[i]
  yield from sub(0)
```

```python
list(queens(8))
```

```python
def queen(n):
  field = [[' 0 '] * n for i in range(n)]
  j = 0 if n % 2 else 1
  
  for i in range(n):
      field[i][j] = ' Ф '
      j = (j + 2) % n
      if n % 2 == 0 and i == (n // 2) - 1:
          j -= 1
  
  for row in field:
      print(''.join(row))
```
