1. Даны две последовательности - арифметическая и геометрическая, первый элемент a, шаг d и q. Найти суммы n первых членов

```python
entered_list = input("Введите a, d, q, n, разделенные пробелом: ").split()
num_list = list(map(float, entered_list))
a = num_list[0]
d = num_list[1]
q = num_list[2]
n = num_list[3]
result1 = a
result2 = a
for a in range(1, n):
  result1 += d
  result2 *= q
print (result1, result2)
```

2. Найти минимальное расстояние между двумя точками в квадрате [(0, 0), (1, 1)]. Периодические граничные условия 

```python
import math
```

```python
enteredList = input("Введите x1, y1, x2, y2, разделенные пробелом: ").split()
numList = list(map(float, enteredList))
x1 = numList[0]
y1 = numList[1]
x2 = numList[2]
y2 = numList[3]
minDist = math.sqrt(2)
for d1 in range(1):
  for d2 in range(1):
    for d3 in range(1):
      for d4 in range(1):
        if (d1 != 1) and (d2 != 1) and (d3 != 1) and (d4 != 1):
          dist = math.sqrt(((x1 + d1) - (x2 + d2))**2 + ((y1 + d3) - (y2 + d4))**2)
          minDist = min(minDist, dist)
print(minDist)
```

3. Выведете список простых чисел не превышающих N

```python
n = int(input())
 
number = []
for i in range(0, n + 1):
    number.append(i)

i = 2
while i <= n:
    if (number[i] != 0):
        j = i + i
        while j <= n:
            number[j] = 0
            j = j + i
    i += 1

number = list(set(number))
numder = number.sort()
print(number)
```

Сравнить время выполнения операции: 
 1. деления на 2 и умножения на 0.5
 2. деления на 4 и умножения на 0.25
 3. np.sqrt и a**0.5
 4. ln и exp
 5. for (0, N-1) итерирующийся по лиcту и через range, изучить зависимость от N

```python
import timeit
```

```python
def f(x):
    return x/2
def g(x):
    return x*0.5

if timeit.timeit('f(10000000)', globals=globals()) > timeit.timeit('g(10000000)', globals=globals()):
  print('* faster then /')
else:
  print('/ faster then *') 
```

```python
def f(x):
    return x/4
def g(x):
    return x*0.25

if timeit.timeit('f(10000000)', globals=globals()) > timeit.timeit('g(10000000)', globals=globals()):
  print('* faster then /')
else:
  print('/ faster then *') 
```

```python
import numpy as np
```

```python
def f(x):
    return np.sqrt(x)
def g(x):
    return x**0.5

if timeit.timeit('f(10000000)', globals=globals()) > timeit.timeit('g(10000000)', globals=globals()):
  print('** faster then np.sqrt')
else:
  print('np.sqrt faster then **') 
```

```python
def f(x):
    return math.log(x)
def g(x):
    return math.exp(x)

if timeit.timeit('f(100)', globals=globals()) > timeit.timeit('g(100)', globals=globals()):
  print('exp faster then ln')
else:
  print('ln faster then exp') 
```
