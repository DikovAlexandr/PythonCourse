```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
H = 0.05
TAU = 0.05
X = np.arange(0, 1.05, H)
Xdim = len(X)
T = np.arange(0, 1.05, TAU)
Tdim = len(T)
U = np.loadtxt("solve.csv")
print(U.size)
```

```python
Uanalytic = np.array([[5/ 2 * np.tanh(x - t) for x in X] for t in T])
```

```python
moment = 20
error = []
for i in range(len(X)):
    error.append(U[moment][i] - Uanalytic[moment][i])

plt.figure(figsize=(10, 10))
plt.plot(X, U[moment], label='FEM')
plt.plot(X, Uanalytic[moment], label='Analytic')
plt.plot(X, error, label='Error')
plt.legend()
plt.show()
```
